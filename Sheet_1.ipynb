**Create a 3 × 3 NumPy ndarray containing consecutive integers from 1 to 9 (inclusive)**

`np.arange()` returns evenly spaced values within a given interval.
- `arange(start_val, stop_val, step_size)`
    - `np.arange(N)` will create a rank 1 ndarray with consecutive integers between `0` and `N - 1`
    - `np.arange(start,stop)` will create a rank 1 ndarray with evenly spaced values within the half-open interval `[start, stop)`. This means the evenly spaced numbers will include `start` but exclude `stop`.
    - `np.arange(start,stop,step)` will create a rank 1 ndarray with evenly spaced values within the half-open interval `[start, stop)` with `step` being the distance between two adjacent values.

```python
import numpy as np
```

```python
X = np.arange(1, 10)
```

```python
X
```

`np.reshape()` gives a new shape to an array without changing its data.
- It is important to note that the new_shape should be compatible with the number of elements in the given ndarray.

```python
X = X.reshape(3,3)
```

```python
X
```

## 10. Create another 2D array **arr2D_2 = np.array([[10, 20, 30, 40], [50, 60, 70, 80]])**, then vertically stack both of the arrays.

- `numpy.hstack(sequence_of_ndarray)`: It returns a stacked array formed by stacking the given arrays in sequence horizontally ***(column-wise)***
- `numpy.vstack(sequence_of_ndarray)`: It returns a stacked array formed by stacking the given arrays, will be at least 2-D, in sequence vertically ***(row-wise)***

**It is important to note that in order to stack ndarrays, the shape of the ndarrays must match.**

```python
arr2D_2 = np.array([
    [10, 20, 30, 40],
     [50, 60, 70, 80]
    ])
```

```python
X = np.vstack([X, arr2D_2])
X
```

## 11. Create a copy of the original array, then slice this copy to extract the first two rows and last two columns.

### **Slicing**
1. `ndarray[start:end]`: is used to select elements between the **'start'** and **'end'** indices.
2. `ndarray[start:]`: is used to select all elements from the **'start'** index till the last index.
3. `ndarray[:end]`: is used to select all elements from the first index till the **'end'** index.

**Note:** in methods one and three, the **'end'** index is excluded.

```python
arr2D_copy = X.copy()
arr2D_copy
```

```python
                    # [row, column]
arr2D_copy = arr2D_copy[:2, -2:]
arr2D_copy
```

# Question 3

![Alt text](https://www.w3resource.com/w3r_images/pandas-data-structure.svg)

```python
import pandas as pd
```

```python
books = ['Great Expectations', 'Of Mice and Men', 'Romeo and Juliet', 'The Time Machine', 'Alice in Wonderland']
authors = ['Charles Dickens', 'John Steinbeck', 'William Shakespeare', 'H. G. Wells', 'Lewis Carroll']

# User ratings (aligned with the book titles above)
user_1 = pd.Series([3.2, 4.0, 2.5])
user_2 = pd.Series([5.0, 1.3, 4.0, 3.8])
user_3 = pd.Series([2.0, 2.3, 3.2, 4.0])
user_4 = pd.Series([4.0, 3.5, 4.0, 5.0, 4.2])
```

## 1. Create a Pandas Series from **books** and **authors** list.

```python
books_series = pd.Series(books)
authors_series = pd.Series(authors)
```

## 2. Create a dictionary using the given data where:
- The keys are **'Author'**, **'Book Title'**, **'User 1'**, **'User 2'**, **'User 3'**, and **'User 4'**.
- The values are the corresponding Series objects.


```python
dat = {'Author': authors_series,
       'Book Title': books_series,
       'User 1': user_1,
       'User 2': user_2,
       'User 3': user_3,
       'User 4': user_4}
```

## 3. Create a Pandas DataFrame from the dictionary, allowing Pandas to automatically assign numerical row indices.

```python
book_ratings = pd.DataFrame(dat)
```

```python
book_ratings
```

## 4. Print key attributes of the DataFrame: The **shape**, **number of dimensions**, **total elements**, **values**, **index**, and **column names**.

```python
print("Shape: ", book_ratings.shape)
print("Number of Dimensions: ", book_ratings.ndim)
print("Total elements: ", book_ratings.size)
print("The data: ", book_ratings.values)
print("The index: ", book_ratings.index)
print("The column index in book ratings: ", book_ratings.columns)
```

## 5. Display only the **'Book Title'** and **'Author'** columns.

```python
print(book_ratings[['Book Title', 'Author']])
```

## 6. Display the first three rows of the DataFrame.

```python
print(book_ratings[:3])
```

```python
book_ratings.head(3)
```

## 7. Insert a new column **'Average Rating'** that represents the average rating across all users for each book.

```python
book_ratings['Average Rating'] = book_ratings[['User 1', 'User 2', 'User 3', 'User 4']].mean(axis=1)
```

```python
book_ratings
```

## 8. Append a new row for the book **"Pride and Prejudice"** by **"Jane Austen"**, with ratings **[4.5, 4.0, 3.8, 4.9]**

```python
new_row = pd.DataFrame([{'Author': "Jane Austen",
       'Book Title': "Pride and Prejudice",
       'User 1': 4.5,
       'User 2': 4.0,
       'User 3': 3.8,
       'User 4': 4.9}])
```

```python
book_ratings = pd.concat([book_ratings, new_row], ignore_index=True)
```

```python
book_ratings
```

## 9. Rename the column **'Book Title'** to **'Title'**

```python
book_ratings.rename(columns={'Book Title': 'Title'}, inplace=True)
```
