```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the news website (example: BBC News)
url = "https://www.bbc.com/news"

# Fetch the page content
response = requests.get(url)
if response.status_code != 200:
    print("Failed to retrieve the webpage")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract headlines (modify the class/tag based on the website structure)
headlines = soup.find_all("h2")
print(headlines)

# Display the top headlines
print("Latest News Headlines:")
for i in range(0,10):  # Limit to first 10 headlines
    print(f"{i+1}. {headlines[i].get_text()}")
```

## Web Scraping Data Retrieval Pt. 2

```python
mport requests
from bs4 import BeautifulSoup
import pandas as pd
URL = "https://wuzzuf.net/search/jobs/?q=data%20analyst&a=hpb" 
response = requests.get(URL)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    job_list = []
    for job_card in soup.find_all("div", class_="css-ghe2tq e1v1l3u10"):
        job_title = job_card.find("a", class_="css-o171kl" ).text.strip() if job_card.find("a", class_="css-o171kl") else "N/A"
        company = job_card.find("a", class_="css-ipsyv7").text.strip() if job_card.find("a", class_="css-ipsyv7") else "N/A"
        location = job_card.find("span", class_="css-16x61xq").text.strip() if job_card.find("span", class_="css-16x61xq") else "N/A"
        contract = job_card.find("span", class_="css-uc9rga eoyjyou0").text.strip() if job_card.find("span", class_="css-uc9rga eoyjyou0") else "N/A"
        job_site = job_card.find("span", class_="css-uofntu eoyjyou0").text.strip() if job_card.find("span", class_="css-uofntu eoyjyou0").text.strip() else "N/A"
        # Store the extracted data
        job_list.append({
            "Job Title": job_title,
            "Company": company,
            "Location": location,
            "Contract": contract,
            "Site": job_site
        })
    # Convert the list into a DataFrame
    df = pd.DataFrame(job_list)
    df.to_csv("job_listings.csv", index=False) # Save the data to a CSV file
    print("\nJob Listings Scraped Successfully!")
    print(df.head())
else:
    print("\nFailed to retrieve the webpage. Status Code:", response.status_code)
```

# **Question 4**
## API Data Retrieval

```python
import requests
import numpy as np
import pandas as pd


# Request weather data in JSON format
URL = f"https://wttr.in/Cairo?format=j1"

response = requests.get(URL)

if response.status_code == 200:
    data = response.json()  # Parse JSON response

    # Extract hourly forecast (Next 24 hours from first day's forecast)
    hourly_forecast = data["weather"][0]["hourly"]

    # Create a NumPy 2D array to store time, temperature, humidity
    forecast_array = np.empty((8, 3), dtype=object)
    for i, hour in enumerate(hourly_forecast):
        time = hour["time"]
        temperature = hour["tempC"]
        humidity = hour["humidity"]
        forecast_array[i] = np.array([[time, temperature, humidity]])

    # Convert NumPy array to Pandas DataFrame
    df = pd.DataFrame(forecast_array, columns=["Time (HHMM)", "Temperature (C)", "Humidity (%)"])

    # Display DataFrame
    print("Hourly Forecast (Next 24 Hours):")
    print(df)

else:
    print("Failed to retrieve data. Try a different city name.")
```
