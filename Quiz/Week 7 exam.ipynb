## 15.	Fill missing values:
## •	Use the median to fill missing values in numerical columns.

- **`fillna(desired_value)`**: Replaces **NaN** values in each column with the desired values.
- **`df[numerical_columns_list].median()`**: computes the median of each numerical column separately.

```python
number_columns = numerical_df.columns
numerical_df[number_columns] = numerical_df[number_columns].fillna(numerical_df[number_columns].median())
```

## 16.	Apply the Interquartile Range (IQR) method for outlier detection on numerical features: •	Print the count of outliers for each numerical column.

```python
for col in numerical_df.columns.tolist():
  Q1 = numerical_df[col].quantile(0.25)
  Q3 = numerical_df[col].quantile(0.75)
  IQR = Q3-Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = numerical_df[(numerical_df[col] < lower_bound) | (numerical_df[col] > upper_bound)]
  print(f"Count of outliers in {col}: {outliers.shape[0]}")
```
